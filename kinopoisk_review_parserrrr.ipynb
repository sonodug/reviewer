```python
import json
import requests
import numpy as np
import pandas as pd
import time
```

добавить дохуя токенов

```python
tokens = ['0197e459-29ee-445b-a733-99ad0ff0020e', '83c9b96f-597a-4fcc-8f97-ffe981e40742']
```

```python
film_data = []

headers = {'X-API-KEY': f'{tokens[0]}', "Content-Type": "application/json"}
```

```python
pages = list(range(1, 20))

for page_num in pages:
  url = f"https://kinopoiskapiunofficial.tech/api/v2.2/films?ratingFrom=1&ratingTo=6&type=FILM&yearFrom=2000&yearTo=2007&page={page_num}"
    
  req = requests.get(url, headers=headers)
  req_result = req.json()

  try:
    film_data.append(req_result['items'])
  except KeyError:
    print('')
```

```python
id_array = []

for data in film_data:
  df = pd.DataFrame.from_records(data)
  id_col = df.loc[:, ['kinopoiskId']]
  id_array.extend(id_col.values)
```

```python
film_ids = np.array(id_array).flatten()
```
